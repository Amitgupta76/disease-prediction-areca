```python
# Function to convert seconds to minutes
def seconds_to_minutes(seconds):
    return seconds / 60
```

```python
# Convert total time to minutes
total_time_rnn_minutes = seconds_to_minutes(total_time_rnn)
total_time_cnn_minutes = seconds_to_minutes(total_time_cnn)
total_time_resnet20_minutes = seconds_to_minutes(total_time_resnet20)
total_time_resnet44_minutes = seconds_to_minutes(total_time_resnet44)

# Calculate average time per epoch (in minutes)
avg_time_rnn = total_time_rnn_minutes / num_epochs_rnn
avg_time_cnn = total_time_cnn_minutes / num_epochs_cnn
avg_time_resnet20 = total_time_resnet20_minutes / num_epochs_resnet20
avg_time_resnet44 = total_time_resnet44_minutes / num_epochs_resnet44
```

```python
# Models
models = ['RNN', 'CNN', 'ResNet20', 'ResNet44']
avg_times = [avg_time_rnn, avg_time_cnn, avg_time_resnet20, avg_time_resnet44]
total_times = [total_time_rnn_minutes, total_time_cnn_minutes, total_time_resnet20_minutes, total_time_resnet44_minutes]
```

```python
import pandas as pd

# Data
models = ['RNN', 'CNN', 'ResNet20', 'ResNet44']
accuracy = [0.963447630405426, 0.9192238450050354, 0.9566786885261536, 0.9711191058158875]
precision = [0.972972972972973, 0.967741935483871, 1.0, 1.0]
recall = [0.972972972972973, 0.967741935483871, 1.0, 1.0]
f1_score = [0.963447630405426, 0.9192238450050354, 0.9566786885261536, 0.9711191058158875]
avg_time_per_epoch = [5.6, 11.89, 34.27, 47.51]
total_time = [134.42, 225.93, 651.08, 902.67]

# Create DataFrame
data = {
    'Model': models,
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1_score,
    'Average Time per Epoch (min)': avg_time_per_epoch,
    'Total Time (min)': total_time
}

df = pd.DataFrame(data)

# Apply styling
styled_df = df.style \
    .format({
        'Accuracy': '{:.2%}',
        'Precision': '{:.2%}',
        'Recall': '{:.2%}',
        'F1 Score': '{:.2%}',
        'Average Time per Epoch (min)': '{:.2f}',
        'Total Time (min)': '{:.2f}'
    }) \
    .bar(subset=['Accuracy', 'Precision', 'Recall', 'F1 Score'], color='#5fba7d') \
    .background_gradient(subset=['Average Time per Epoch (min)', 'Total Time (min)'], cmap='Blues')

# Display styled DataFrame
styled_df
```
